```python
import pandas as pd 
data = pd.read_csv("password_strength_data.csv", on_bad_lines='skip')
data.head()
```

0 means: the password is weak
1 means: the password is medium
2 means: the password is strong

```python
print(data.isnull().sum())
data = data.dropna()
```

```python
data["strength"] = data["strength"].map({0: "weak",
                                         1: "Medium",
                                         2: "Strong"})
```

```python
data.sample(5)
```

```python
import numpy as np
```

```python
x = np.array(data["password"])
y = np.array(data["strength"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
def word(password):
    character=[]
    for i in password:
        character.append(i)
    return character
```

```python
tdif = TfidfVectorizer(tokenizer=word)
x = tdif.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, 
                                                test_size=0.05, 
                                                random_state=42)
```

```python
#print(xtrain[1], ytrain[1])
#print(ytest.size)

```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100)
```

Note: It will take some time

```python
model.fit(xtrain, ytrain)
```

```python
print(model.score(xtest, ytest))
```

```python
user_pass = input("Enter Password: ")
pass_tok = tdif.transform([user_pass]).toarray()
print(model.predict(pass_tok))
```
